```python
import pandas as pd
```

```python
# Dataset de personas
personas = pd.DataFrame({
    "ID": ["P1", "P2", "P3", "P4", "P5"],
    "Nombre": ["Laura", "Jorge", "Camila", "Luis", "Ana"],
    "Profesión": ["Ingeniera de SW", "Analista de datos", "UX Designer", "DevOps", "PM Junior"],
    "Años de experiencia": [5, 3, 7, 4, 6],
    "Edad": [32, 28, 35, 30, 33],
    "Ubicación": ["Bogotá", "Medellín", "Cali", "Bogotá", "Medellín"],
    "Habilidades": [
        {"Python", "Git", "SQL", "Liderazgo"},
        {"Excel", "SQL", "Tableau", "Comunicación"},
        {"Figma", "Prototipado", "Comunicación"},
        {"Docker", "AWS", "Python", "Automatización"},
        {"Scrum", "Comunicación", "Jira", "Planificación"}
    ],
    "Disponibilidad (h/día)": [6, 4, 5, 6, 5]
})
```

```python
# Dataset de componentes
componentes = pd.DataFrame({
    "ID Componente": ["C1", "C2", "C3", "C4", "C5"],
    "Nombre": [
        "Panel de visualización",
        "Diseño de interfaz móvil",
        "Backend de autenticación",
        "Despliegue automático",
        "Gestión y planificación"
    ],
    "Habilidades Requeridas": [
        {"SQL", "Tableau", "Comunicación"},
        {"Figma", "Prototipado"},
        {"Python", "Git", "Docker"},
        {"Docker", "AWS", "Automatización"},
        {"Jira", "Scrum", "Comunicación"}
    ],
    "Horas Estimadas": [20, 16, 24, 18, 12],
    "Prioridad": ["Alta", "Media", "Alta", "Media", "Baja"]
})
```

```python
# Función para hacer el matching de habilidades
def match_personas_componentes(personas, componentes, min_habilidades_comunes=2, min_disponibilidad=5):
    asignaciones = []

    for _, comp in componentes.iterrows():
        for _, pers in personas.iterrows():
            if pers["Disponibilidad (h/día)"] >= min_disponibilidad:
                habilidades_comunes = pers["Habilidades"].intersection(comp["Habilidades Requeridas"])
                if len(habilidades_comunes) >= min_habilidades_comunes:
                    asignaciones.append({
                        "Componente": comp["Nombre"],
                        "Persona": pers["Nombre"],
                        "Habilidades Coincidentes": habilidades_comunes,
                        "Total Coincidencias": len(habilidades_comunes),
                        "Prioridad Componente": comp["Prioridad"]
                    })

    return pd.DataFrame(asignaciones)

# Realizar la asignación
asignaciones_df = match_personas_componentes(personas, componentes)
asignaciones_df
```
