```python
import pandas as pd

# Crear un DataFrame con entregables
data = {
    'entregable': ['Función A', 'Función B', 'Función C', 'Validación QA', 'Ciberseguridad'],
    'tipo': ['alcance', 'alcance', 'alcance', 'calidad', 'riesgo'],
    'valor_planificado': [100, 150, 120, 0.9, 30],  # calidad como % (0.9) y riesgo como unidades de reducción
    'costo_planificado': [80, 110, 90, 10, 0]       # opcional, útil si luego se desea calcular CPI
}

df = pd.DataFrame(data)

# Filtrar entregables según tipo
alcance_total = df[df['tipo'] == 'alcance']['valor_planificado'].sum()
factor_calidad = df[df['tipo'] == 'calidad']['valor_planificado'].prod()  # producto de factores de calidad
reduccion_riesgos = df[df['tipo'] == 'riesgo']['valor_planificado'].sum()

# Calcular valor ganado ajustado
valor_esperado = (alcance_total * factor_calidad) - reduccion_riesgos

# Mostrar resultados
print("Alcance total planificado:", alcance_total)
print("Factor de calidad aplicado:", factor_calidad)
print("Reducción por riesgos esperados:", reduccion_riesgos)
print("Valor ganado proyectado (ajustado):", valor_esperado)
```
